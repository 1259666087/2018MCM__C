```python
import pandas as pd
import numpy
import csv
import sys
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
pro=pd.read_csv('C:\\Users\\yyf\\Desktop\\MCM\\TEPRB.csv')
pro
```

```python
pro2=pro.set_index(['StateCode','Year'])
pro2
```

```python
AZ=pro2.ix['AZ']['Data']
CA=pro2.ix['CA']['Data']
NM=pro2.ix['NM']['Data']
TX=pro2.ix['TX']['Data']
AZ
```

```python
fig,axes=plt.subplots(1,1)
axes.plot(AZ,label='Arizona')
axes.plot(CA,label='California')
axes.plot(NM,label='New Mexico')
axes.plot(TX,label='Texas')
axes.set_title('Total Energy production')
axes.set_xlabel('Year')
axes.grid(True)
axes.set_ylabel('Energy/(Billion Btu)')
plt.legend(loc = 'upper right') 
```
